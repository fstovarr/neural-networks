```r
rm(list=ls())

library(dplyr)
library(data.table)
library(tidyr)
library(ggplot2)
library(grid)
library(gridExtra)
library(purrr)
library(corrplot)

setwd("C:/Users/u0500/Downloads/archive")

Data.P=fread("Datos.csv",dec = ",")
Data.P=Data.P[,-1]

str(Data.P)

Data.F=separate(Data.P, Nombre, c("Genus","Especie","ID"), sep="-")

# Data.F[,.(mean_fft=mean(fft_var),Med_fft=median(fft_var),max_fft=max(fft_var),min_fft=min(fft_var)), by="Especie"]

crear_boxplot <- function(Data.F, nombre){
  
  if(nombre %like% "_min") b <- "de los valores min"
  if(nombre %like% "_max") b <- "de los valores max"
  if(nombre %like% "_mean") b <- "de los valores promedio"
  if(nombre %like% "_var") b <- "de las varianzas"
  if(nombre %like% "x_") a <- " de la frecuencia \n del audio"
  if(nombre %like% "spectral_") a <- " del espectro \n del audio"
  if(nombre %like% "fft") a <- " de la transformada de \n fourier del audio"
  
  p1 <- ggplot(Data.F, aes(x=Especie, y=get(nombre), fill=Especie)) +
    geom_boxplot() + theme(legend.position="none") + 
    scale_y_continuous(name=nombre) +
    theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 18),
          axis.text.y = element_text(angle = 0, hjust = 1, size = 22),
          text = element_text(size=22) ,
          plot.title = element_text(hjust = 0.5,size = 26, face = "bold")) + 
          ggtitle(paste0("Boxplot ",b,a," con outliers"))
  tiff(filename=paste("Boxplots/","Boxplot ",nombre,"con outliers.png"),width = 1000, height = 500)
  grid.arrange(p1,nrow=1)
  dev.off()
  
  p2 <- ggplot(Data.F, aes(x=Especie, y=get(nombre), fill=Especie)) +
    geom_boxplot(outlier.shape=NA) + theme(legend.position="none") + 
    scale_y_continuous(name=nombre, limits=quantile(Data.F[,get(nombre)], c(0.1,0.9))) +
    theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 18),
          axis.text.y = element_text(angle = 0, hjust = 1, size = 22),
          text = element_text(size=22) ,
          plot.title = element_text(hjust = 0.5,size = 26, face = "bold")) + 
    ggtitle(paste0("Boxplot ",b,a," sin outliers"))
  tiff(filename=paste("Boxplots/","Boxplot ",nombre," sin outliers.png"),width = 1000, height = 500)
  grid.arrange(p2,nrow=1)
  dev.off()
}

nombres <- names(Data.F)
nombres <- nombres[!(nombres %in% c("Especie","Genus","ID"))]

nombres %>% map(~crear_boxplot(Data.F=Data.F,nombre=.x))

str(Data.F)

names(Data.F)
componentes_cli<-prcomp(Data.F[,1:11], scale=TRUE,center = TRUE)
ls(componentes_cli)
dev.new()
barplot(componentes_cli$sdev)

data.plot <- data.frame(componentes_cli$x[,c("PC1","PC2")])
p <- ggplot(data.plot, aes(x=PC1,y=PC2)) +
  geom_point(aes(colour=Data.F$Especie))

p



cor.mtest <- function(mat, ...) {
  mat <- as.matrix(mat)
  n <- ncol(mat)
  p.mat<- matrix(NA, n, n)
  diag(p.mat) <- 0
  for (i in 1:(n - 1)) {
    for (j in (i + 1):n) {
      tmp <- cor.test(mat[, i], mat[, j], ...)
      p.mat[i, j] <- p.mat[j, i] <- tmp$p.value
    }
  }
  colnames(p.mat) <- rownames(p.mat) <- colnames(mat)
  return(p.mat)
}

Graf_corr <- function(Data){
  
  M <- cor(Data)
  
  p.mat <- cor.mtest(M)
  
  title <- ""
  col <- colorRampPalette(c("#BB4444", "#EE9988", "#FFFFFF", "#77AADD", "#4477AA"))
  p <- corrplot(M, method="square", col=col(200),  
                diag=FALSE, # tl.pos="d",
                type="upper", # order="hclust",
                title=title,tl.col = "black",
                #addCoef.col = "black", # Add coefficient of correlation
                # Combine with significance
                p.mat = p.mat,  sig.level = 1, insig = "blank",
                # hide correlation coefficient on the principal diagonal
                mar=c(0,0,2,0) # http://stackoverflow.com/a/14754408/54964
  )
  return(p)
}


Data.F[,ID:=ave(rep(1,nrow(Data.F)),Data.F$Especie, FUN=seq_along)]
spectro <- Data.F %>% select("ID","Especie","spectral_mean") %>%
            spread(Especie, value="spectral_mean")

tiff(filename=paste("Boxplots/correlacion.png"),width = 1200, height = 700)
Graf_corr(spectro)
dev.off()



data <- fread("Ruidos.csv", dec=",")

```
